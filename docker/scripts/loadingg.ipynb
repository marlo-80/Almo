```python
import os
import shutil
import time
import kagglehub
import sys

KAGGLE_PATH = "robikscube/flight-delay-dataset-20182022"
OUTPUT_DIR = "./flight_data"          # Passe ggf. den Pfad an

# Zielverzeichnis anlegen
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Prüfen, ob schon CSV-Dateien da sind
existing = [f for f in os.listdir(OUTPUT_DIR) if f.startswith("Combined_Flights_") and f.endswith(".csv")]
if existing:
    print(f"CSV-Dateien bereits vorhanden: {existing}. Überspringe Download.")
    sys.exit(0)

# Download mit Wiederholungen
MAX_RETRIES = 5
for attempt in range(1, MAX_RETRIES + 1):
    try:
        print(f"Download-Versuch {attempt}/{MAX_RETRIES} ...")
        # KEIN output_dir – kompatibel mit 0.3.10
        cache_path = kagglehub.dataset_download(KAGGLE_PATH)
        print(f"Download abgeschlossen in: {cache_path}")

        # Alle CSV-Dateien in OUTPUT_DIR kopieren
        for file in os.listdir(cache_path):
            if file.endswith(".csv"):
                src = os.path.join(cache_path, file)
                dst = os.path.join(OUTPUT_DIR, file)
                shutil.copy2(src, dst)
                print(f"Kopiert: {file}")
        break  # Erfolg

    except Exception as e:
        print(f"Fehler bei Versuch {attempt}: {e}")
        # Teilweise geladene Cache-Dateien löschen
        try:
            shutil.rmtree(cache_path, ignore_errors=True)
        except NameError:
            pass
        if attempt == MAX_RETRIES:
            raise RuntimeError("Download nach 5 Versuchen abgebrochen.")
        time.sleep(10)  # 10 Sekunden Pause

print("✅ Fertig! CSV-Dateien liegen jetzt in", OUTPUT_DIR)
```

```python
import os
import shutil
import time
import kagglehub
import sys
import resource

# 1. Open‑File‑Limit erhöhen (für Linux/macOS)
try:
    soft, hard = resource.getrlimit(resource.RLIMIT_NOFILE)
    resource.setrlimit(resource.RLIMIT_NOFILE, (8192, hard))
    print(f"Limit erhöht von {soft} auf 8192")
except Exception as e:
    print(f"Limit-Erhöhung nicht möglich: {e}")

KAGGLE_PATH = "robikscube/flight-delay-dataset-20182022"
OUTPUT_DIR = "./flight_data"          # Passe ggf. an

os.makedirs(OUTPUT_DIR, exist_ok=True)

# Prüfen, ob schon CSV-Dateien da sind – dann überspringen
existing = [f for f in os.listdir(OUTPUT_DIR) if f.startswith("Combined_Flights_") and f.endswith(".csv")]
if existing:
    print(f"CSV-Dateien bereits vorhanden: {existing}. Überspringe Download.")
    sys.exit(0)

# Download mit Wiederholungen
MAX_RETRIES = 5
for attempt in range(1, MAX_RETRIES + 1):
    try:
        print(f"Download-Versuch {attempt}/{MAX_RETRIES} ...")
        # KEIN output_dir – kompatibel mit 0.3.10
        cache_path = kagglehub.dataset_download(KAGGLE_PATH)
        print(f"Download abgeschlossen in: {cache_path}")

        # Alle CSV-Dateien in OUTPUT_DIR kopieren
        for file in os.listdir(cache_path):
            if file.endswith(".csv"):
                src = os.path.join(cache_path, file)
                dst = os.path.join(OUTPUT_DIR, file)
                shutil.copy2(src, dst)
                print(f"Kopiert: {file}")
        break  # Erfolg

    except Exception as e:
        print(f"Fehler bei Versuch {attempt}: {e}")
        # Cache leeren (teilweise Dateien)
        try:
            # Falls cache_path definiert ist, lösche ihn
            if 'cache_path' in locals() and os.path.exists(cache_path):
                shutil.rmtree(cache_path, ignore_errors=True)
        except:
            pass
        if attempt == MAX_RETRIES:
            raise RuntimeError("Download nach 5 Versuchen abgebrochen.")
        time.sleep(10)  # 10 Sekunden Pause

print("✅ Fertig! CSV-Dateien liegen jetzt in", OUTPUT_DIR)
```
